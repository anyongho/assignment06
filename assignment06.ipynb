## import packages for plotting graphs and manipulate data
### And write matplotlib incline for plot error

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline

num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b
```

## this is for how to measure the least squre fit method (y = kx + m)

```python
meanx = sum(x) / len(x)
meany = sum(y1) / len(y1)
k = sum((xi-meanx)*(yi-meany) for xi,yi in zip(x,y1)) / sum((xi-meanx)**2 for xi in x)
m = meany - k*meanx

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```

## plot the line of least square

```python
y = k * x + m
plt.plot(x,y,'r')
```

## plot the line of least square and points

```python
line = plt.plot(x,y)
plt.setp(line, color = 'r', linewidth = 5.0)
plt.plot(x, y1, 'b.', x, y2, 'k--')
plt.show()
```
